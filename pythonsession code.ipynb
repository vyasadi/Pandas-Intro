```python
# Read csv
data=pd.read_csv('samplepython.csv',skipfooter=2,engine='python')
```

```python
# New calculation
data['CTR']=data['Clicks']/data['Impressions']
data
```

```python
# convert Date column to datetime 
data['Date']=data['Date'].apply(pd.to_datetime)
data.info()
```

```python
# Filter data for clicks below 25000
dataclicksbelow25000=data[data['Clicks']<25000]
dataclicksbelow25000
```

```python
# Filter data for category as pants
datapants=data[data['Category']=='Pants']
datapants
```

```python
# multiple filter
dataclicksbelow15000andcategorypants=data[(data['Clicks']<15000)&(data['Category']=='Pants')]
dataclicksbelow15000andcategorypants
```

```python
# Group by and aggregation
aggregateddata=data.groupby(['Date','Category']).agg({'Clicks' : 'sum','Cost' : 'mean'})
aggregateddata
```

```python
# Stacking
data.groupby(['Date','Category']).agg({'Click' : 'sum','Cost' : 'mean'}).unstack('Category')
```
